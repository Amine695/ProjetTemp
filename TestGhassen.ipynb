```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

import numpy as np
import math

z= np.linspace(0,20,100)#axe vertical descendant
t= np.linspace(0,8760,100)#axe de temps

T=np.zeros(len(z))
L=0.8
rho=1800
c=835
w=2*10**(-7)
a=L/(rho*c)

#T(z=0,t)=Ti+T0cos(wt)
Ti=4 #température moyenne au pifpour le test
T0=5  #température maximale au pifpour le test

for i in range(len(z)):
    for j in range(len(t)):
        T[i] = (Ti+T0*math.exp(-math.sqrt(w/(2*a))*z[i])*np.cos(w*t[j]-math.sqrt(w/(2*a))*z[i]))
```

```python
fig=plt.figure() #T(z,t)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(z, t, T, c='r', marker='o')
ax.set_xlabel('Profondeur')
ax.set_ylabel('Temps')
ax.set_zlabel('Températures')
plt.show()
```
