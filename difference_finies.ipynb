```python
import numpy as np
from matplotlib.pyplot import *
from scipy.sparse import diags  # Pour les matrices diagonales


T_0 = 10 # température au niveau du sol (ici valeur aléatoire pour tester les fonctions)

# Les variables qui suivent sont de type float
# Valeur None pour l'instant car on doit faire des recherches

pC=None # capacité thermique volumique du sol
L=None # conductivité thermique du sol

#C=L/(pC)  # constante
C=1.0

# Différences finies

Tsol = 10  #condition initiale
Tlimite = 0   #condition finale


# Pour le moment, on étudie sur 10 points mais il faudrait faire un while qui s'arrête quand les deux dernières valeurs sont trop proches, avec une certaine précision p
N=10

temps=100 # on étudie l'évolution sur 100 secondes



# On discrétise l'intervalle [0,1] en N+1 noeuds de coordonnées z_i (i variant de 0 à N)
z=np.linspace(0,1,N+1)  
dz =z[1]-z[0]  # on calcule le pas de z soit la distance entre deux points de z (car z a été créé avec linspace du coup points équidistants)

#pas_z=dz

dt=0.1 # intervalle de temps en s, on choisit une petite valeur pour ne pas fausser les résultats

def calcul_lambda(C,dt,dz):
    return (C*dt)/(dz**2)

lbda=calcul_lambda(1,dt,dz)

# Création de la matrice tridiagonale

# Taille N-1 pour les prochaines matrices

def creation_matrice_tridiagonale(lbda,N): #lbda pour lambda et N, nombre de points étudiés
    k = np.array([-lbda*np.ones(N-2),1+2*lbda*np.ones(N-1),-lbda*np.ones(N-2)])
    offset = [-1,0,1]
    A = diags(k,offset).toarray()
    return A

A=creation_matrice_tridiagonale(lbda,N)

def creation_vecteur_conditions_limites(Tsol,Tlimite,N):
    T_zeros=np.zeros(N-1) 
    T_zeros[0]=Tsol
    T_zeros[-1]=Tlimite
    return T_zeros

T_zeros=creation_vecteur_conditions_limites(Tsol,Tlimite,N)
#print('T_zeros',T_zeros)


# Le nombre d'itérations n vaut temp/dt
ite=int(temps/dt)



# Résolution système d'équations linéaires 
# T_n_suiv=np.linalg.solve(A,T_n+lbda*T_zeros) (à chaque itération de la boucle for)



# Schéma IMPLICITE de la discrétisation de l'équation de la chaleur en 1D

def implicite(A,lbda,T_zeros,nb_iterations,N):
    T_n_suiv=np.zeros(N-1) # T à l'étape n+1 (<=> au temps (n+1)*dt) et initialisé à 0 ici
    
    for i in range(nb_iterations):
        T_n_suiv=np.linalg.solve(A,T_n_suiv+lbda*T_zeros)
        
        
        #print('T =',T_n_suiv)
    
    return T_n_suiv



# Schéma EXPLICITE de la discrétisation de l'équation de la chaleur en 1D

def explicite(A,lbda,T_zeros,nb_iterations,N):
    T_n_suiv=np.ones(N-1) # T à l'étape n+1 (<=> au temps (n+1)*dt) et initialisé à 0 ici
    T_n_suiv[0]=Tsol
    T_n_suiv[-1]=Tlimite
    #T_n_suiv=T_zeros[0]*np.ones(N-1)
    
    
    
    for i in range(nb_iterations):
        T_n_suiv=(A.dot(T_n_suiv))+lbda*T_zeros
        
        
        # Autre formule possible
        #T_n_suiv=(np.eye(N-1)-lbda*A).dot(T_n_suiv)
        
        
        
        
        #print('T =',T_n_suiv)
        
    return T_n_suiv
    

T_implicite=implicite(A,lbda,T_zeros,ite,N)

T_explicite=explicite(A,lbda,T_zeros,3,N)

print('T_im',T_implicite)
print('T_ex',T_explicite)

    
```
